## Graficar 2.36

La función a graficar es:

$$
f(x) = 
  \left \{
    \begin{aligned}
      &2-x &0< x < 4\\
      &x-6 &4 < x < 8
    \end{aligned}
  \right .
$$

graficaremos la función y su aproximación con la serie de Fourier:

```python
# Definimos la función que vamos a aproximar

def f_2_36(x):
    
    if ( 0<=x and x<4 ):
        return 2-x
    else:
        return x-6

x = np.linspace(0, 8, 100) # Arreglo de 0 a 8 en 100 pasos

y = [f_2_36(i) for i in x] # Imágenes de los valores del arreglo de x

# Ahora graficamos

plt.figure()

plt.plot(x,y)
plt. plot(x, [0 for i in x])
plt.xlabel("x")
plt.ylabel("y") # Cambiamos las etiquetas de los ejes
plt.legend(["f(x)", "Eje x"])

plt.show()
```

### Serie de Fourier

La expansión es la siguiente:

$$
\begin{aligned}
      f(x) &= \frac{16}{\pi^2}\sum_{n=1}^\infty\frac{1}{(2n-1)^2}\cos\frac{(2n-1)\pi x}{4}\\\\
      \Rightarrow f(x) &= \frac{16}{\pi^2} \left[ \frac{\cos(\pi x)}{4} + \frac{1}{3^2} \frac{\cos(3\pi x)}{4} + \frac{1}{5^2} \frac{\cos(5\pi x)}{4} + \cdots \right]
\end{aligned}
$$

```python
# Creamos la función que nos devuelve los valores en el eje "y" de la aproximación para un n dado

def ApproxFourier(x,n):
    
    sumatoria = 0
    
    for i in range(1,n+1):
        sumatoria += (1/np.power((2*i-1),2)) * np.cos(((2*i-1) * np.pi * x)/4)

    return 16/np.power(np.pi,2) * sumatoria



y_fou_n1 = ApproxFourier(x, 1)    # Las imagenes de la aproximación con n = 1
y_fou_n3 = ApproxFourier(x, 3)    # Las imagenes de la aproximación con n = 3
y_fou_n10 = ApproxFourier(x, 10)  # Las imagenes de la aproximación con n = 10


# Ahora graficamos

plt.figure()

plt.plot(x,y, linewidth=4) # Graficamos la funcion

plt.plot(x, y_fou_n1)  # Graficamos la aproximación con n = 1
plt.plot(x, y_fou_n3)  # Graficamos la aproximación con n = 3
plt.plot(x, y_fou_n10) # Graficamos la aproximación con n = 10

plt.xlabel("x")
plt.ylabel("y") # Cambiamos las etiquetas de los ejes
plt.legend(["f(x)= x(pi - x)", "n = 1", "n = 3", "n = 10"]) # Leyenda de la gráfica

plt.show()
```

## Graficar 2.40

La función a graficar es:

$$
f(x) = x(\pi - x)
$$

para $0 \leq x \leq \pi$,

graficaremos la función y su aproximación con serie de senos y en otra gráfica su aproximación con serie de cosenos:

```python
# Definimos la función que vamos a aproximar

def f_2_40(x):
    
    return x * (np.pi - x)

x = np.linspace(0, np.pi, 100) # Arreglo de 0 a pi en 100 pasos

y = f_2_40(x) # Imágenes de los valores del arreglo de x

# Ahora graficamos

plt.figure()

plt.plot(x,y)
plt. plot(x, [0 for i in x])
plt.xlabel("x")
plt.ylabel("y") # Cambiamos las etiquetas de los ejes
plt.legend(["f(x)= x(pi - x)", "Eje x"])

plt.show()
```

### (a) Serie de Cosenos

La expansión es la siguiente:

$$
x(\pi-x) = \frac{\pi^2}{6} - \left[ \frac{\cos(2x)}{1^2} + \frac{\cos(4x)}{2^2} + \frac{\cos(6x)}{3^2} + \cdots \right]
$$

```python
# Creamos la función que nos devuelve los valores en el eje "y" de la aproximación para un n dado

def ApproxCosenos(x,n):
    
    sumatoria = 0
    
    for i in range(1,n+1):
        sumatoria += np.cos(2*i*x)/np.power(i,2)

    return np.power(np.pi,2)/6 - sumatoria



y_cos_n1 = ApproxCosenos(x, 1)    # Las imagenes de la aproximación con n = 1
y_cos_n3 = ApproxCosenos(x, 3)    # Las imagenes de la aproximación con n = 3
y_cos_n10 = ApproxCosenos(x, 10)  # Las imagenes de la aproximación con n = 10


# Ahora graficamos

plt.figure()

plt.plot(x,y, linewidth=4) # Graficamos la funcion

plt.plot(x,y_cos_n1)  # Graficamos la aproximación con n = 1
plt.plot(x,y_cos_n3)  # Graficamos la aproximación con n = 3
plt.plot(x,y_cos_n10) # Graficamos la aproximación con n = 10

plt.xlabel("x")
plt.ylabel("y") # Cambiamos las etiquetas de los ejes
plt.legend(["f(x)= x(pi - x)", "n = 1", "n = 3", "n = 10"]) # Leyenda de la gráfica

plt.show()
```

Hacemos un acercamiento a la parte de arriba para apreciar mejor la diferencia

```python
plt.figure()

plt.plot(x,y, linewidth=4) # Graficamos la funcion

plt.plot(x,y_cos_n1)  # Graficamos la aproximación con n = 1
plt.plot(x,y_cos_n3)  # Graficamos la aproximación con n = 3
plt.plot(x,y_cos_n10) # Graficamos la aproximación con n = 10

plt.xlabel("x")
plt.ylabel("y") # Cambiamos las etiquetas de los ejes
plt.legend(["f(x)= x(pi - x)", "n = 1", "n = 3", "n = 10"]) # Leyenda de la gráfica

plt.xlim([0.85, 2.25]) # Cambiamos los límites para hacer el acercamiento
plt.ylim([2, 2.7])

plt.show()
```

### (a) Serie de Senos

La expansión es la siguiente:

$$
x(\pi-x) = \frac{8}{\pi} \left[ \frac{\sin(x)}{1^3} + \frac{\sin(3x)}{3^3} + \frac{\sin(5x)}{5^3} + \cdots \right]
$$

```python
# Creamos la función que nos devuelve los valores en el eje "y" de la aproximación para un n dado

def ApproxSenos(x,n):
    
    sumatoria = 0
    
    for i in range(1,n+1):
        sumatoria += np.sin( (2*i - 1)*x ) / np.power(2*i - 1, 3)

    return 8/np.pi * sumatoria



y_sin_n1 = ApproxSenos(x, 1)    # Las imagenes de la aproximación con n = 1
y_sin_n3 = ApproxSenos(x, 3)    # Las imagenes de la aproximación con n = 3
y_sin_n10 = ApproxSenos(x, 10)  # Las imagenes de la aproximación con n = 10


# Ahora graficamos

plt.figure()

plt.plot(x, y, linewidth=4) # Graficamos la funcion

plt.plot(x, y_sin_n1)  # Graficamos la aproximación con n = 1
plt.plot(x, y_sin_n3)  # Graficamos la aproximación con n = 3
plt.plot(x, y_sin_n10) # Graficamos la aproximación con n = 10

plt.xlabel("x")
plt.ylabel("y") # Cambiamos las etiquetas de los ejes
plt.legend(["f(x)= x(pi - x)", "n = 1", "n = 3", "n = 10"]) # Leyenda de la gráfica

plt.show()
```

# Calcular numéricamente $\pi$ usando el ejercicio 2.46

### (a)

La igualdad que usaremos es la de la parte (a) del ejercicio 2.46 del Spiegel:

$$
\sum_{n=1}^\infty \frac{1}{n^4} = \frac{\pi^4}{90}
$$

$$
\Rightarrow \pi = \sqrt[4]{90 ( \frac{1}{1^4} + \frac{1}{2^4} + \frac{1}{3^4} \cdots)}
$$

Vamos a crear una función que reciba el número de $n$ y que devuelva el valor aproximado de $\pi$ con dicho $n$

```python
def ApproxPiPartA(n):
    
    sumatoria = 0

    for i in range(1, n+1):
        sumatoria += 1/np.power(i,4)

    return np.power( 90*sumatoria, 1/4)
```

Ahora se llama la función para $n=3$, $n=5$ y $n=10$

```python
ApproxPiPartA(3)
```

```python
ApproxPiPartA(5)
```

```python
ApproxPiPartA(10)
```

### (b)

La igualdad que usaremos es la de la parte (b) del ejercicio 2.46 del Spiegel:

$$
\frac{\pi^6}{960} =  \sum_{n=1, n \, impar}^\infty \frac{1}{(n)^6}
$$

$$
\Rightarrow \pi = \sqrt[6]{960 ( \frac{1}{1^6} + \frac{1}{3^6} + \frac{1}{5^6} \cdots)}
$$

```python
def ApproxPiPartB(n):
    
    sumatoria = 0

    for i in range(1, n+1):
        sumatoria += 1/np.power( 2*i - 1,6)

    return np.power( 960*sumatoria, 1/6)
```

Ahora se llama la función para $n=3$, $n=5$ y $n=10$
